```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

'''
Generar variables del universo

   * Variables con rangos subjetivos
   
   11 Inputs (1-Fiebre, 2-Dolorcabeza, 3-Doloresgenerales, 4-Fatiga, 5-Nariztapada
              6-Estornudos, 7-Gargantairritada, 8-Tos, 9-Dificultadrespirar,
              10-Escurrimientonazal, 11-Diarrea)

    1 Output (1-Diagnostico)

'''
# 11 Inputs
x_fiebre = np.arange(35, 42, 1)
x_dolcav = np.arange(0, 11, 1)
x_dolgral = np.arange(0, 11, 1)
x_fatiga = np.arange(0, 11, 1)
#x_nariztap = np.arange(0, 11, 1)
#x_estornud = np.arange(0, 11, 1)
#x_gargirri = np.arange(0, 11, 1) 
#x_tos = np.arange(0, 11, 1)
#x_difresp = np.arange(0, 11, 1)
#x_escurrnaz = np.arange(0, 11, 1)
#x_diarrea = np.arange(0, 11, 1)

# 1 Output
x_diag  = np.arange(0, 11, 1)

# Generando funciones de membresa difusas

# 1-Fiebre  (Inputs)
fiebre_no = fuzz.trimf(x_fiebre, [33, 35, 37])
fiebre_si = fuzz.trimf(x_fiebre, [34, 36, 38])

# 2-Dolorcabeza
dolcav_no = fuzz.trimf(x_dolcav, [0, 0, 5])
dolcav_bajo = fuzz.trimf(x_dolcav, [0, 5, 10])
dolcav_intenso = fuzz.trimf(x_dolcav, [5, 10, 10])

# 3-DolorGeneral
dolgral_no = fuzz.trimf(x_dolgral, [0, 0, 2])
dolgral_leve = fuzz.trimf(x_dolgral, [1, 3, 6])
dolgral_severo = fuzz.trimf(x_dolgral, [5, 10, 10])

# 4-Fatiga
fatiga_no = fuzz.trimf(x_fatiga, [0, 0, 2])
fatiga_leve = fuzz.trimf(x_fatiga, [1, 2, 5])
fatiga_severa = fuzz.trimf(x_fatiga, [2, 6, 8])
fatiga_extrema = fuzz.trimf(x_fatiga, [7, 10, 10])

# 1-Diagnostico (Outputs)
diag_1 = fuzz.trimf(x_diag, [0, 0, 3])
diag_2 = fuzz.trimf(x_diag, [0, 2, 5])
diag_3 = fuzz.trimf(x_diag, [2, 5, 8])
diag_4 = fuzz.trimf(x_diag, [4, 7, 10])
diag_5 = fuzz.trimf(x_diag, [7, 10, 13])

# Visualizacin para los universos y funciones de pertenencia
# ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10, ax11

fig, (ax1, ax2, ax3, ax4, ax12) = plt.subplots(nrows=5, figsize=(10, 15))

# Funciones de membresia Inputs (graficos)

# Fiebre
ax1.plot(x_fiebre, fiebre_no, 'b', linewidth=1.5, label='No')
ax1.plot(x_fiebre, fiebre_si, 'g', linewidth=1.5, label='Bajo')
ax1.set_xlim([35, 41])
ax1.set_title('1-Fiebre')
ax1.legend()

# Dolor de cabeza
ax2.plot(x_dolcav, dolcav_no, 'b', linewidth=1.5, label='No')
ax2.plot(x_dolcav, dolcav_bajo, 'g', linewidth=1.5, label='Bajo')
ax2.plot(x_dolcav, dolcav_intenso, 'r', linewidth=1.5, label='Intenso')
ax2.set_title('2-Dolor de cabeza')
ax2.legend()

# Dolor General
ax3.plot(x_dolgral, dolgral_no, 'b', linewidth=1.5, label='No')
ax3.plot(x_dolgral, dolgral_leve, 'g', linewidth=1.5, label='Leves')
ax3.plot(x_dolgral, dolgral_severo, 'r', linewidth=1.5, label='Severos')
ax3.set_title('3-Dolor general')
ax3.legend()

# Fatiga
ax4.plot(x_fatiga, fatiga_no, 'b', linewidth=1.5, label='No')
ax4.plot(x_fatiga, fatiga_leve, 'g', linewidth=1.5, label='Leve')
ax4.plot(x_fatiga, fatiga_severa, 'r', linewidth=1.5, label='Severa')
ax4.plot(x_fatiga, fatiga_extrema, 'y', linewidth=1.5, label='Extrema')
ax4.set_title('4-Fatiga')
ax4.legend()

# Funciones de membresia Outputs (graficos)

#Diagnostico
ax12.plot(x_diag, diag_1, 'b', linewidth=1.5, label='MuyPocoProb')
ax12.plot(x_diag, diag_2, 'g', linewidth=1.5, label='PocoProb')
ax12.plot(x_diag, diag_3, 'r', linewidth=1.5, label='Posible')
ax12.plot(x_diag, diag_4, 'y', linewidth=1.5, label='Probable')
ax12.plot(x_diag, diag_5, 'c', linewidth=1.5, label='MuyProb')
ax12.set_title('Diagnostico (output)')
ax12.legend()

# Desactivar los ejes superior/derecho
for ax in (ax1, ax2, ax3, ax4, ax12):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```
